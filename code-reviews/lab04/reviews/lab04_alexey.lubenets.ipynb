```python
import json
import pandas as pd
import numpy as np
```

```python
file_path = '/data/share/lab04data/'

train_file = file_path + 'lab04_train.csv'
test_file = file_path + 'lab04_test.csv'
```

```python
train_df = pd.read_csv(train_file, sep=',')
print(train_df.shape)
train_df[:10].T
```

```python
test_df = pd.read_csv(test_file, sep=',')
print(test_df.shape)
test_df[:10].T
```

```python
print(train_df.loc[~ train_df['TARGET'].isin([1,0])].T)
train_df = train_df[train_df['TARGET'].notna()]
```

# Числовые фичи

```python
train_df['AGE_IN_YEARS'] = train_df['AGE']//12
test_df['AGE_IN_YEARS'] = test_df['AGE']//12

train_df['MAX_PCLOSE_DATE_IN_YEAR'] = train_df['MAX_PCLOSE_DATE']//12
test_df['MAX_PCLOSE_DATE_IN_YEARS'] = test_df['MAX_PCLOSE_DATE']//12

num_cols_0 = [
    'APP_REGISTR_RGN_CODE',
]

num_cols = [
    'AGE_IN_YEARS', 'AGE',
    'MAX_PCLOSE_DATE_IN_YEAR', 'MAX_PCLOSE_DATE',
    'AMOUNT_RUB_CLO_PRC','AMOUNT_RUB_NAS_PRC','AMOUNT_RUB_SUP_PRC','AMOUNT_RUB_ATM_PRC',
    
    'CLNT_SETUP_TENOR',
    
    'CNT_TRAN_CLO_TENDENCY1M','CNT_TRAN_AUT_TENDENCY1M','CNT_TRAN_AUT_TENDENCY3M','CNT_TRAN_MED_TENDENCY1M',
    'CNT_TRAN_MED_TENDENCY3M','CNT_TRAN_SUP_TENDENCY1M','CNT_TRAN_ATM_TENDENCY3M','CNT_TRAN_ATM_TENDENCY1M',
    'CNT_TRAN_CLO_TENDENCY3M','CNT_TRAN_SUP_TENDENCY3M',
    
    'DEAL_GRACE_DAYS_ACC_MAX','DEAL_GRACE_DAYS_ACC_AVG','DEAL_GRACE_DAYS_ACC_S1X1',
    
    'DEAL_YWZ_IR_MIN','DEAL_YWZ_IR_MAX',
    
    'CR_PROD_CNT_CCFP','CR_PROD_CNT_CC','CR_PROD_CNT_TOVR','CR_PROD_CNT_PIL','CR_PROD_CNT_VCU','CR_PROD_CNT_IL',
    
    'LDEAL_ACT_DAYS_PCT_TR3','LDEAL_ACT_DAYS_PCT_AAVG','LDEAL_ACT_DAYS_PCT_CURR',
    
    'LDEAL_DELINQ_PER_MAXYWZ','LDEAL_GRACE_DAYS_PCT_MED','LDEAL_USED_AMT_AVG_YWZ','LDEAL_ACT_DAYS_ACC_PCT_AVG',
    
    'TURNOVER_DYNAMIC_IL_1M','TURNOVER_DYNAMIC_CC_3M',
    
    'REST_DYNAMIC_CC_3M','MED_DEBT_PRC_YWZ',
    
    'SUM_TRAN_MED_TENDENCY1M','SUM_TRAN_AUT_TENDENCY1M','SUM_TRAN_AUT_TENDENCY3M','SUM_TRAN_CLO_TENDENCY1M',
    'REST_DYNAMIC_SAVE_3M','REST_DYNAMIC_FDEP_1M','REST_AVG_CUR',
    'TRANS_COUNT_NAS_PRC','TRANS_COUNT_SUP_PRC',
    
    'TURNOVER_CC','TRANS_COUNT_ATM_PRC','TURNOVER_PAYM',
    'SUM_TRAN_MED_TENDENCY3M','REST_DYNAMIC_FDEP_3M','REST_DYNAMIC_IL_1M',
    'TURNOVER_DYNAMIC_CC_1M',
    'TRANS_AMOUNT_TENDENCY3M','TRANS_CNT_TENDENCY3M',
    'REST_DYNAMIC_CC_1M','TURNOVER_DYNAMIC_PAYM_3M',
    'SUM_TRAN_SUP_TENDENCY1M','SUM_TRAN_ATM_TENDENCY1M',
    'TURNOVER_DYNAMIC_PAYM_1M','TURNOVER_DYNAMIC_CUR_3M',
    'REST_DYNAMIC_PAYM_1M','TURNOVER_DYNAMIC_IL_3M','SUM_TRAN_ATM_TENDENCY3M',
    'REST_DYNAMIC_PAYM_3M','SUM_TRAN_SUP_TENDENCY3M','REST_DYNAMIC_IL_3M',
    'SUM_TRAN_CLO_TENDENCY3M',
    'REST_DYNAMIC_CUR_1M','REST_AVG_PAYM',
    'REST_DYNAMIC_CUR_3M','TURNOVER_DYNAMIC_CUR_1M',
]

num_low_cols = [
# < 1000 cnt in target=1
    'AVG_PCT_DEBT_TO_DEAL_AMT','AVG_PCT_MONTH_TO_PCLOSE',
    'CLNT_SALARY_VALUE',
    
    'LDEAL_AMT_MONTH','LDEAL_YQZ_PC',
    # + categorial
    'LDEAL_YQZ_CHRG','LDEAL_YQZ_COM',
    # < 1200 cnt in target=1
    'DEAL_YQZ_IR_MAX','DEAL_YQZ_IR_MIN',
    
    'LDEAL_TENOR_MAX','LDEAL_TENOR_MIN',
    'MED_DEBT_PRC_YQZ',
    # + categorial
    'LDEAL_USED_AMT_AVG_YQZ','LDEAL_DELINQ_PER_MAXYQZ',
    
    # eq 0     
#     'Unnamed: 0','ID',
    'CNT_ACCEPTS_TK','CNT_ACCEPTS_MTP',
    'LDEAL_ACT_DAYS_PCT_TR','LDEAL_ACT_DAYS_PCT_TR4',
    
    'PRC_ACCEPTS_A_ATM','PRC_ACCEPTS_MTP','PRC_ACCEPTS_TK','PRC_ACCEPTS_A_MTP',
    'PRC_ACCEPTS_A_AMOBILE','PRC_ACCEPTS_A_TK','PRC_ACCEPTS_A_TK','PRC_ACCEPTS_A_POS','PRC_ACCEPTS_A_EMAIL_LINK',
    
]

# заменяем пробелы в классических числовых полях на средние
for field in num_cols + num_low_cols:
    print(field)
    train_df[field] = train_df[field].fillna(train_df[field].mean())
    test_df[field] = test_df[field].fillna(test_df[field].mean())

# заменяем пробелы в регионах на 0
for field in num_cols_0:
    print(field)
    train_df[field] = train_df[field].fillna(0)
    test_df[field] = test_df[field].fillna(0)
```

# Категориальные фичи

```python
categorial = [
#     из этой фичи не смог достать пользный эффект
#     'CLNT_JOB_POSITION',
    'APP_MARITAL_STATUS', 'APP_COMP_TYPE', 'APP_KIND_OF_PROP_HABITATION', 'APP_EDUCATION',
    'APP_EMP_TYPE', 'APP_POSITION_TYPE',
    
    'CLNT_JOB_POSITION_TYPE','CLNT_TRUST_RELATION',
    
    'PACK',    
]

# приводим к нижнему регистру
for field in categorial:
    print(field)
    train_df[field] = train_df[field].str.lower()
    test_df[field] = test_df[field].str.lower()

train_df[categorial].T
```

```python
# создаем one encoding категориальных переменных
df = pd.get_dummies(train_df[categorial])
tdf = pd.get_dummies(test_df[categorial])

df[:10].T
```

```python
# вытаскиваем из выборок только те категории, которые имеются в обоих наборах
cat_cols = list(set.intersection(set(df.columns.tolist()),set(tdf.columns.tolist())))
print(len(cat_cols))
print(cat_cols)

train_df = train_df.join(df[cat_cols])
test_df = test_df.join(tdf[cat_cols])
```

```python
binary_cols = [
    'APP_DRIVING_LICENSE', 'APP_TRAVEL_PASS', 'APP_CAR'
]

train_df[binary_cols].T
```

```python
# приведение к требуемому виду двоичных признаков
for df in [train_df, test_df]:
    for field in binary_cols:
        print(field)
        df.loc[df[field].isnull(), field] = 0
        df.loc[df[field] == 'Y', field] = 2
        df.loc[df[field] == 'N', field] = 1

train_df[binary_cols][:100].T
test_df[binary_cols][:100].T
```

## Training

```python
# формирование итогового списка фичей для обучения
train_cols = num_cols + num_cols_0 + num_low_cols + binary_cols + cat_cols
    
print(len(train_cols))
print(train_cols)

X = train_df[train_cols]
Y = train_df['TARGET']
```

## RandomForest

```python
# поиск оптимального решения через поиск по сетке, победил - RFC
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold, GridSearchCV

pipe = Pipeline([('classifier' , RandomForestClassifier())])

# Create param grid.
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

param_grid = [
#     {
#         'classifier' : [LogisticRegression()],
#         'classifier__penalty' : ['l1', 'l2'],
#         'classifier__C' : np.logspace(-4, 4, 20),
#         'classifier__solver' : ['liblinear']},
    {
        'classifier' : [RandomForestClassifier()],
        'classifier__n_estimators' : list([120]),
#         'classifier__max_features' : list(range(6,32,5))
        'classifier__max_features' : list([30,15])
    }
]

# Create grid search object
clf = GridSearchCV(pipe, param_grid = param_grid, cv=cv, verbose=True, n_jobs=-1)

# Fit on data
best_clf = clf.fit(X, Y)
```

```python
print(f"Best score: {best_clf.best_score_}")

best_model = best_clf.best_estimator_

print("Best parameters set:")
best_parameters = best_clf.best_estimator_.get_params()
for param_name in sorted(best_parameters.keys()):
    print(f"\t{param_name}: {best_parameters[param_name]}")
```

# PREDICT

```python
# фактический предикт проба
print('predict_proba')

model = best_clf
res = model.predict_proba(test_df[train_cols])

res = pd.DataFrame(res, columns = ['A', 'B'])

test_df['target'] = res['B']
test_df[['ID','target']]
```
