```python
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 7, 100).astype(np.float32)
x += np.random.randn(100)
```

```python
y = 3.5 * x + 1.5
# y = 2.5 * x ** 2 + 1.5 * x + 0.5
# y = 15 * np.sin(x)
```

```python
y += np.random.randn(100) * 2
```

```python
y /= 100
```

```python
plt.plot(x, y, 'o')
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(1),
])
```

```python
model.compile(optimizer='sgd', loss='mse')
```

```python
model.fit(x[..., None], y[..., None], epochs=100)
```

```python
x_test = np.linspace(-5, 10, 200).astype(np.float32)
y_pred = model(x_test[..., None]).numpy()[..., 0]
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x, y, 'o')
plt.plot(x_test, y_pred)
```
