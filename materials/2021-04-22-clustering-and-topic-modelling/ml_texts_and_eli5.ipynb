```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv', sep=';')
```

```python
df = df[df['target'].isin([0,1])]
```

```python
from bs4 import BeautifulSoup
```

```python
from tqdm import tqdm
texts = [BeautifulSoup(x).text for x in tqdm(df['text'])]
```

```python
import pymystem3
m = pymystem3.Mystem()
import tqdm

import re
lem_texts = []
for i in tqdm.tqdm(texts):
    lem_texts.append( [x for x in m.lemmatize(i) if re.match(u'[а-яё]+', x) and len(x) >= 4] )
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

m = cv.fit_transform([' '.join(x) for x in lem_texts])
```

```python
pd.DataFrame(m.toarray())
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(m, df['target'])
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_
```

```python
pd.DataFrame(
    zip(
        rf.feature_importances_,
        cv.get_feature_names()
    )
).sort_values(0, ascending=False)
```
